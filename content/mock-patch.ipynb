What are mocks?
What is patching?
- IO. Main examples:
 - Requests
 - DB
 - Files
 - `input()`

How to patch with Python
- Decorator (on methods and classes)
 - With an extra argument in unit test to pass the mock to the body of the test e.g. `@patch('') def test_something(self, db_connect_mock): ...`
 - Without an extra arg to define the mock outside of the body of the test e.g. `@patch('', MagicMocks) def test_something(self): ...`
- Context manager
- Functions in before and after

What is the path that I should mock?

- Patch the function/object that was imported by the thing being tested

Writing clean mocks

Use attribute chaining instead of defining lots of individual mocks e.g.
```
execute_mock = MagicMock()

db_connection_mock.return_value \
    .cursor.return_value \
    .execute = execute_mock
```
Instead of:
```
connection_mock = MagicMock()
cursor_mock = MagicMock()
cursor_function_mock = MagicMock()
connection_mock.cursor = cursor_function_mock
cursor_function_mock.return_value = cursor_mock

execute_mock = MagicMock()
connection_mock.commit = commit_mock
```
Avoid nesting context managers:
```
with patch('whatever') as whatever_mock:
    with patch('another_thing') as another_thing_mock:
        with patch('again_something') as again_something_mock:
            # do some testing here
```

Instead:

```
with patch('whatever') as whatever_mock, \
    patch('another_thing') as another_thing_mock, \
    patch('again_something') as again_something_mock:
    # Do some testing here
```

A better way is to use dependency injection:
- Separate out the IO from the business logic
- Fewer things to patch at once, less patch setup when testing business logic
